```python
# importing python libraries NumPy, pandas, Matplotlib and seaborn 
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt  
sns.set_style('darkgrid') # to suppress warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
#To get top 5 results of the dataframe
df = pd.read_csv("globalterrorismdb_0718dist.csv",encoding='latin1')
df.head()
```

```python
#to get the values of all the columns
df.columns.values
```

```python
#To get information about the dataframe
df.info()
```

```python
#To view some basic statistical details like percentile, mean, std etc. of a data frame
df.describe()
```

```python
#To remove any column with all Nan Value
df.dropna(how='all',axis=1)
df.head()
```

```python
#To rename the existing coulmns
df.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)


```

```python
# filtering the important coulmns
df = df[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
#To find the top five values of new dataframe
df.head()
```

```python
#To find the total number of null value in the column
df.isnull().sum() 
```

```python
#To plot total number of attacks VS year
year = df['Year'].unique()
years_count = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 90)
plt.xlabel(' Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Total number of attacks in years',fontsize=40)
plt.show()
```

# **Maximum number of attacks in year=2014**

```python
#To plot total number of people killed VS year
df1 = df[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(16,10))
df1.plot(kind='bar',alpha=1,ax=ax4)
plt.xticks(rotation = 90)
plt.title("People Died in Attack",fontsize=25)
plt.ylabel("Number of people killed",fontsize=20)
plt.xlabel('Year',fontsize=20)
```

```python
#To plot total number of people wounded VS year
df1 = df[['Year','Wound']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(16,10))
df1.plot(kind='bar',alpha=1,ax=ax4)
plt.xticks(rotation = 90)
plt.title("People wounded in Attack",fontsize=40)
plt.ylabel("Number of people wounded",fontsize=20)
plt.xlabel('Year',fontsize=20)
```

```python
#To count the top 5 country which faced maximum attacks
attack = df.Country.value_counts()[:5]
attack
```

```python
#To plot total number of attacks VS country
plt.subplots(figsize=(15,6))
country_attacks = df.Country.value_counts()[:15].reset_index()
country_attacks.columns = ['Country', 'Total Attacks']
sns.barplot(x=country_attacks.Country, y=country_attacks['Total Attacks'], palette="mako",edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number Of Total Attacks in Each Country')
plt.show()
```

```python
#To count the top 5 city which faced maximum attacks
attack = df.City.value_counts()[:5]
attack
```

```python
#To plot total number of attacks VS city
df['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).drop('Unknown').head(15).plot(kind='bar',figsize=(30,15),color='red')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 15 most effected city",fontsize=20)
plt.show()
```

```python
#To count the regions which faced maximum attacks
attack = df.Region.value_counts()[:15]
attack
```

```python
#To plot total number of attacks VS region
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(30,15))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
#To plot total number of attacks VS attack_type
df['Attacktype'].value_counts().to_frame().sort_values('Attacktype',axis=0,ascending=False).drop('Unknown').head(15).plot(kind='bar',figsize=(20,15),color='red')
plt.xticks(rotation = 90)
plt.xlabel("Attack_type",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.title("Attack_type Count",fontsize=30)
plt.show()
```

```python
#To plot total number of attacks VS groups
df['Group'].value_counts().to_frame().drop('Unknown')[:20].plot(kind='bar',color='red',figsize=(20,10))
plt.title("Top 20 terrorist group attack",fontsize=40)
plt.xlabel("terrorist group",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```
