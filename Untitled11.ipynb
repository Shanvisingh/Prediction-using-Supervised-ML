```python
# importing python libraries NumPy, pandas, Matplotlib and seaborn 
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt  
sns.set_style('darkgrid') # to suppress warnings 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("globalterrorismdb_0718dist.csv",encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#To remove any column with all Nan Value
df.dropna(how='all',axis=1)
df.head()
```

```python
df.info()
```

```python
#rename the important columns
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)


```

```python
# filtering the important coulmns
df = df[['Year','Month','Day','Country','state','Region','city','latitude','longitude',"AttackType",'Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
#To find the top five values of new dataframe
df.head()
```

```python
#To find the total number of null value in the column
df.isnull().sum() 
```

```python
year = df['Year'].unique()
years_count = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 90)
plt.xlabel(' Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.title('Total number of attacks in years',fontsize=40)
plt.show()
```

```python
df1 = df[['Year','Killed']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(16,10))
df1.plot(kind='bar',alpha=1,ax=ax4)
plt.xticks(rotation = 90)
plt.title("People Died in Attack",fontsize=25)
plt.ylabel("Number of people killed",fontsize=20)
plt.xlabel('Year',fontsize=20)
```

```python
df1 = df[['Year','Wounded']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(16,10))
df1.plot(kind='bar',alpha=1,ax=ax4)
plt.xticks(rotation = 90)
plt.title("People wounded in Attack",fontsize=40)
plt.ylabel("Number of people wounded",fontsize=20)
plt.xlabel('Year',fontsize=20)
```

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(20,15))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
attack_type = df.AttackType.value_counts().to_frame().reset_index()
attack_type.columns = ['Attack Type', 'Total Attacks']
plt.subplots(figsize=(15,6))
sns.barplot(x=attack_type['Attack Type'], y=attack_type['Total Attacks'], palette='rocket',
            edgecolor=sns.color_palette('dark', 10))
plt.xticks(rotation=90)
plt.title('Attack Type Count')
plt.show()
```
